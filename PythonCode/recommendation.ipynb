```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
import pandas as pd
data = pd.read_csv('../Dataset/preprocessed_dataset.csv')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
vectors_df = pd.DataFrame(vectors,  columns=cv.get_feature_names_out())
```

```python
print(vectors_df)
```

```python
# Save processed dataset
vectors_df.to_csv("../Dataset/vectors.csv", index=False)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(product_id):
    idx = data[data['Product_id'] == product_id].index[0]
    print(idx)
    print(data.iloc[idx].URL)
    distances = similarity[idx]
    data_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in data_list:
        print(data.iloc[i[0]].URL)
```

```python
recommend(11690882)
```

```python

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import nltk
from nltk.stem.porter import PorterStemmer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
import joblib


knn = NearestNeighbors(n_neighbors=10, metric='cosine', algorithm='brute')
knn.fit(vectors)

# Save the KNN model locally
joblib.dump(knn, 'knn_model.pkl')
```

```python
import joblib

# Load the model from the file
knn_model = joblib.load('knn_model.pkl')

sample_index = 1
distances, indices = knn_model.kneighbors([vectors[sample_index]])

print(f"Nearest neighbors for item at index {sample_index}:")
print("Indices:", indices)
print("Distances:", distances[0])
```
